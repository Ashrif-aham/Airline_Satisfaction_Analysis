Exploratory Data Analysis(EDA)

```python
sns.countplot(x='satisfaction',data=df)
plt.title("Customer Satisfaction Distribution")
plt.show()
```

```python
cat_cols = ["Gender","Customer_Type","Type_of_Travel","Class"]
```

```python
fig,axes = plt.subplots(2,2,figsize=(12,10))
axes = axes.flatten()

for i,col in enumerate(cat_cols):
    sns.countplot(x=col,hue='satisfaction', data=df, ax=axes[i])
    axes[i].set_title(f"{col} vs Satisfaction")
    axes[i].tick_params(axis='x',rotation=30)

plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['Age'],bins=30,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['Flight_Distance'],bins=30,kde=True)
plt.title("Flight Distance")
plt.show()
```

```python
sns.boxplot(x='satisfaction',y='Departure_Delay_in_Minutes',data=df)
plt.title("Departure_Delay")
plt.show()
```

```python
sns.boxplot(x='satisfaction',y='Arrival_Delay_in_Minutes',data=df)
plt.title("Arrival_Delay")
plt.show()
```

```python
service_cols=["Seat_comfort","Food_and_drink","Online_support","Cleanliness","On-board_service"]

fig,axes = plt.subplots(2,3,figsize=(15,10))
axes = axes.flatten()

for i,col in enumerate(service_cols):
    sns.boxplot(x="satisfaction",y=col,data=df,ax=axes[i])
    axes[i].set_title(f"{col} vs Satisfaction")


fig.delaxes(axes[5])

plt.tight_layout()
plt.show()
```

Correlation Analysis
